```
import pandas as pd
grades = pd.read_excel("Marks.xlsx",sheet_name="Grades", skiprows=3)
grades
```

```
#drop the empty rows
grades_cleaned = grades.dropna(how="all")
grades_cleaned
```

```
#grades replace NA values with 0
grades_cleaned = grades_cleaned.fillna(0)
grades_cleaned
```

```
#Change the index to ID permanently and display the data frame
grades_cleaned.set_index("ID ", inplace=True)
grades_cleaned
```

```
#Reset the index back to original permanently. Verify it by display the data frame.
grades_cleaned.reset_index(inplace=True)
grades_cleaned.head(3)
```

```
#Display the average FWA(Final) marks
grades_cleaned["Final"].mean()
```

```
#Display the Statistics Summary for the column Project
grades_cleaned["Project"].describe()
```

```
#Create a new column 'Final Marks' which is sum of 20% of Quiz, 25% of Practical, 25% of Project and 30% of FWA. Verify the above command to display the data in grades
grades_cleaned["Final Marks"] = (0.2 * grades_cleaned["Quiz"]) + (0.25 * grades_cleaned["Practical"]) + (0.25 * grades_cleaned["Project"]) + (0.3 * grades_cleaned["Final"]) 
grades_cleaned.head()
```

```
#display the highest value in final marks
grades_cleaned["Final Marks"].max() 
```
